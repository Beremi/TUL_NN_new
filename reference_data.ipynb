```python
import pandas as pd
import numpy as np

# Load the CSV file, using the first column from the CSV as the DataFrame index.
df = pd.read_csv('mesured_data_time_axis.csv', index_col=False)

# Display the first few rows of the corrected DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['time'], df['V1'], label='V1')
plt.plot(df['time'], df['H1'], label='H1')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('V1 and H1 vs Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import h5py

# Stack V1 and H1 as columns to form a 2D array
v1_h1 = df[['V1', 'H1']].values.T.flatten()
plt.plot(v1_h1)
```

```python

# Save to HDF5 file
with h5py.File('v1_h1_data.h5', 'w') as f:
    f.create_dataset('v1_h1', data=v1_h1)
```
